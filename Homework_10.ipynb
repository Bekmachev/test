1. В столбце Name всего 2 строки с отсутствующими значениями, для удобства работы удалим эти строки.
2. В столбце Year_of_Release видно, что вылеты отсутствуют, поэтому заменим отсутствующие значения на среднее значение и округлим год до целого.
3. Отсутствующие значения в столбце Genre удалились при удалении строк в п. 1.
4. Отсутствующие значения в столбце Publisher можно заменить на самого частовстречаемого издателя или прописать "Other", корректнее использовать "Other".
5. В столбцах Critic_Score, Critic_Count, User_Score, User_Count процент отсутствующих значений составляют существенную часть данных в датасете (40-54%), заменим отсутствующие значения в них на средние значения (вылетов этих столбцах нет), только перед этим столбец User_Score надо привести в численный формат.
6. Значения в столбце Developer для многих игр совпадает со значением в столбце Publisher, поэтому заменим отсутствующие значения в стоблце Developer на значения в столбце Publisher/
7. Отсутствующие значения в столбце Rating заменим на 'Unknown', ввиду отсутствия какого-либо сильно преобладающего значения.

```python
df_games.dropna(subset=['Name'], inplace=True)
```

```python
df_games['Year_of_Release'].fillna(df_games['Year_of_Release'].mean(), inplace=True)
df_games['Year_of_Release'] = df_games['Year_of_Release'].astype(int)
```

```python
df_games['Publisher'].value_counts()
```

```python
df_games['Publisher'].fillna('Other', inplace=True)
```

```python
df_games['User_Score'].value_counts()
```

tbd - вероятно означает to be determined = будет определено, что также можно считать отсутствующим значением, поэтому заменим его тоже на среднее значение в столбце

```python
def del_chart(x):
    if x == 'tbd': x = None
    return x
df_games['User_Score'] = df_games['User_Score'].apply(del_chart)
```

```python
df_games['Critic_Score'].fillna(df_games['Critic_Score'].mean(), inplace=True)

df_games['Critic_Count'].fillna(df_games['Critic_Count'].mean(), inplace=True)
df_games['Critic_Count'] = df_games['Critic_Count'].astype(int)

df_games['User_Score'] = df_games['User_Score'].astype(float)
df_games['User_Score'].fillna(df_games['User_Score'].mean(), inplace=True)

df_games['User_Count'].fillna(df_games['User_Count'].mean(), inplace=True)
df_games['User_Count'] = df_games['User_Count'].astype(int)
```

```python
df_games['Developer'].fillna(df_games['Publisher'], inplace=True)
```

```python
df_games['Rating'].value_counts()
```

```python
df_games['Rating'].fillna('Unknown', inplace=True)
```
