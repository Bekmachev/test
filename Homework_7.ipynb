**Ввиду того, что в рассмотренном ранее датасете отсутсвуют пропуски, загрузим новый датасет с сайта** https://www.kaggle.com/datasets/alitaqi000/world-university-rankings-2023

Стоит отметить, что большая часть строк с пропусками находится внизу таблицы, ниже строки 1697, и данные по этим университетеам еще не заполнены. Если выводить только данные выше этой строки, то количество пропусков получется следующее:

```python
data_reyt_univ.isnull().loc[:1697].sum()
```
