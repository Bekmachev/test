**Выполним нормализацию численных данных**

Большинство данных в таблице уже отнормированы, за исключением столбца No of student, однако применив функцию info(), видно, что данные в столбец имеют строковыйти тип, поэтому сначала преобразуем данные в численный вид. Заодно преобразуем столбец International Student в целочисленный вид

```python
data_reyt_univ.info()
```

```python
def change_chart(y):
    y = str(y).replace(',', '')
    y = float(y)
    return y
data_reyt_univ['No of student'] = data_reyt_univ['No of student'].apply(change_chart)
data_reyt_univ.sort_values(by = 'No of student', ascending=True)
```

```python
def del_chart(x):
    x = str(x).replace('%', '')
    if x == '': x = None
    else: x = float(x)
    return x
data_reyt_univ['International Student'] = data_reyt_univ['International Student'].apply(del_chart)
data_reyt_univ.sort_values(by = 'International Student', ascending=False)
```

```python
normalized_data = normalize(data_reyt_univ[['No of student', 'No of student per staff']].loc[:1500], norm='l1', axis=1)
normalized_data
```

**Посчитаем процент пропусков в таблице**

```python
data_reyt_univ.isnull().sum() * 100 / len(data_reyt_univ)
```

1. Строки без названия университета можно удалить ввиду их малого количества и потому что без названия университета остальные значения не несут практической информации (после удаления значительно сократился процент пропусков в столбцах, содержащих информацию о студентах)
2. Пропуски в названии локации заполним строкой "Unknown"
3. Пропуски в колонках No of student, No of student per staff, International Student, Teaching Score,	Research Score,	Citations Score,	Industry Income Score,	International Outlook Score заменим средними значениями, поскольку в датасете отсутствуют выбросы
4. Пропуски в столбцах Female:Male Ratio и OverAll Score заменим на "Unknown"

```python
data_reyt_univ.dropna(subset=['Name of University'], inplace=True)
```

```python
data_reyt_univ['Location'].fillna('Unknown', inplace=True)
```

```python
data_reyt_univ['No of student'].fillna(data_reyt_univ['No of student'].mean(), inplace=True)
data_reyt_univ['No of student per staff'].fillna(data_reyt_univ['No of student per staff'].mean(), inplace=True)
data_reyt_univ['International Student'].fillna(data_reyt_univ['International Student'].mean(), inplace=True)
data_reyt_univ['Teaching Score'].fillna(data_reyt_univ['Teaching Score'].mean(), inplace=True)
data_reyt_univ['Research Score'].fillna(data_reyt_univ['Research Score'].mean(), inplace=True)
data_reyt_univ['Citations Score'].fillna(data_reyt_univ['Citations Score'].mean(), inplace=True)
data_reyt_univ['Industry Income Score'].fillna(data_reyt_univ['Industry Income Score'].mean(), inplace=True)
data_reyt_univ['International Outlook Score'].fillna(data_reyt_univ['International Outlook Score'].mean(), inplace=True)
```

```python
data_reyt_univ['Female:Male Ratio'].fillna('Unknown', inplace=True)
```

```python
data_reyt_univ['OverAll Score'].fillna('Unknown', inplace=True)
```
